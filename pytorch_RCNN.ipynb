```python
import torch
import torchvision
from torchvision import transforms as T
from PIL import Image
import cv2
```

```python
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained = True)
```

```python
model.eval()
```

```python
ig = Image.open("stop.jpg")
```

```python
transform = T.ToTensor()
img = transform(ig)
```

```python
with torch.no_grad():
    pred = model([img])#With our own image
print(pred)#boxes, prediction, saying class und accuracy, es de tipo dict, donde xisten las keys (boxes, labels, score)
```

```python
bboxes, labels, scores = pred[0]["boxes"], pred[0]["labels"], pred[0]["scores"]
#print(scores)
```

```python
#Ahora se tomaran los mayores a 90%
num = torch.argwhere(scores > 0.9).shape[0] # vale 5
```

```python
coco_names = ["person" , "bicycle" , "car" , "motorcycle" , "airplane" , "bus" , "train" , "truck" , "boat" , "traffic light" , "fire hydrant" , "street sign" , "stop sign" , "parking meter" , "bench" , "bird" , "cat" , "dog" , "horse" , "sheep" , "cow" , "elephant" , "bear" , "zebra" , "giraffe" , "hat" , "backpack" , "umbrella" , "shoe" , "eye glasses" , "handbag" , "tie" , "suitcase" , 
"frisbee" , "skis" , "snowboard" , "sports ball" , "kite" , "baseball bat" , 
"baseball glove" , "skateboard" , "surfboard" , "tennis racket" , "bottle" , 
"plate" , "wine glass" , "cup" , "fork" , "knife" , "spoon" , "bowl" , 
"banana" , "apple" , "sandwich" , "orange" , "broccoli" , "carrot" , "hot dog" ,
"pizza" , "donut" , "cake" , "chair" , "couch" , "potted plant" , "bed" ,
"mirror" , "dining table" , "window" , "desk" , "toilet" , "door" , "tv" ,
"laptop" , "mouse" , "remote" , "keyboard" , "cell phone" , "microwave" ,
"oven" , "toaster" , "sink" , "refrigerator" , "blender" , "book" ,
"clock" , "vase" , "scissors" , "teddy bear" , "hair drier" , "toothbrush" , "hair brush"]
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
igg = cv2.imread("stop.jpg")
for i in range(num):
    x1, y1, x2, y2 = bboxes[i].numpy().astype("int")#numpy para que sea en formato de numeros no de tensor, y el int para que sean en enteros
    class_name = coco_names[labels.numpy()[i]-1]#-1 porque es una lista
    class_score = scores.numpy()[i] # obtener precision
    text = f"{class_name}: {class_score:.2f}"
    igg = cv2.rectangle(igg, (x1, y1), (x2, y2), (0, 255, 0), 1)#nombre, las coordenadas y color
    igg = cv2.putText(igg, text, (x1, y1), font, 0.5, (0, 255, 0), 1, cv2.LINE_AA)#incluir dentro de cv
```

```python
cv2.imshow('Imagen',igg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
